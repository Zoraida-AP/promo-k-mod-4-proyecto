Jueves de antaño: Ventas de árboles de Navidad en Estados Unidos

Este conjunto de datos ofrece un registro exhaustivo de las ventas anuales de árboles de Navidad en Estados Unidos desde 2010 hasta 2016. El conjunto de datos consta de seis columnas que incluyen información relevante sobre los datos de ventas de cada año.

La columna Año indica el año específico en el que se registraron los datos de ventas de árboles de Navidad, lo que permite a los analistas comparar y realizar un seguimiento de las tendencias a lo largo del tiempo.

La columna Tipo de árbol especifica las distintas especies o tipos de árboles de Navidad que se vendieron durante cada año concreto, lo que permite a los investigadores analizar las preferencias del mercado y las elecciones de los consumidores.

La columna Número de árboles vendidos representa la cantidad total de árboles de Navidad comprados por los clientes en un año determinado. La identificación de las fluctuaciones de esta métrica puede ofrecer información sobre los cambios en la demanda y el comportamiento del mercado.

La columna Precio medio de los árboles proporciona información importante sobre la dinámica de precios en el sector. Mediante el cálculo y seguimiento de este precio medio para cada año, los analistas pueden evaluar las variaciones en el comportamiento del gasto de los consumidores, así como identificar los posibles factores económicos que influyen en las decisiones de compra.

Por último, la columna Ventas presenta datos valiosos sobre los ingresos totales generados anualmente por estas ventas de árboles de Navidad. Esta métrica ofrece una perspectiva holística sobre el rendimiento del mercado y la rentabilidad empresarial dentro de la temporada navideña.

En conjunto, este detallado conjunto de datos constituye un recurso fiable para los investigadores que pretenden comprender la situación del mercado de los árboles de Navidad.

Esto puede utilizarse para analizar qué tipos específicos de árboles son más populares o generan mayores ingresos por ventas en un año determinado.

```python
def exploracion(df):
    df_info = pd.DataFrame()
    df_info["% nulos"] = round(df.isna().sum()/df.shape[0]*100, 2).astype(str)+"%"
    df_info["% no_nulos"] = round(df.notna().sum()/df.shape[0]*100, 2).astype(str)+"%"
    df_info["tipo_dato"] = df.dtypes
    df_info["num_valores_unicos"] = df.nunique()
    print(f"""El DataFrame tiene {df.shape[0]} filas y {df.shape[1]} columnas.
Tiene {df.duplicated().sum()} datos duplicados, lo que supone un porcentaje de {round(df.duplicated().sum()/df.shape[0], 2)}% de los datos.
Hay {len(list(df_info[(df_info["% nulos"] != "0.0%")].index))} columnas con datos nulos, y son:
{list(df_info[(df_info["% nulos"] != "0.0%")].index)}
y sin nulos hay {len(list(df_info[(df_info["% nulos"] == "0.0%")].index))} columnas y son:
{list(df_info[(df_info["% nulos"] == "0.0%")].index)}
A continuación tienes un detalle sobre los datos nulos y los tipos y número de datos:""")
    display(df_info.head())
    print("Principales estadísticos de las columnas categóricas:")
    display(df.describe(include="O").T)
    print("Principales estadísticos de las columnas numéricas:")
    display(df.describe(exclude="O").T)
    return df_info

exploracion(df)
```

# DUPLICADOS

```python
def get_duplicate_rows(df):
    """
    Encuentra y devuelve las filas completas duplicadas de un DataFrame.

    Args:
        df (pd.DataFrame): El DataFrame de entrada.

    Returns:
        pd.DataFrame: Un nuevo DataFrame con las filas duplicadas completas.
    """
    # Identificar duplicados basados en todas las columnas
    duplicate_rows = df[df.duplicated(keep=False)]
    return duplicate_rows

get_duplicate_rows(df)
```
